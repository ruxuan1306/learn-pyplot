```python
import matplotlib.pyplot as plt
import numpy as np

def f1(t):
    return 2*t

def f2(t):
    return 2 * 1.1**t

x = np.linspace(0, 26, 1000)
y1 = f1(x)
y2 = f2(x)

plt.plot(x, y1, label='y=2t')
plt.plot(x, y2, label='y=2*1.1^t')
plt.axvline(x=24, linestyle='--', color='gray')
plt.axvline(x=26, linestyle='--', color='gray')
plt.legend()
plt.show()
```
